### The beauty of `objects`

```python
def base_figure():
    fig, ax = plt.subplots(figsize=(6,5))
    for index, cyl in enumerate([4,6,8]):
        df = cars[cars['cyl'] == cyl]
        ax.scatter(df['wt'], df['mpg'], c=color[index], s=100, alpha=0.75, edgecolor='none')
    return fig, ax
```

```python
fig, ax = base_figure()

ax.xaxis.set_ticks_position('none')
ax.yaxis.set_ticks_position('none')  

fig.show()
```

```python
def remove_ticks(ax):
    ax.xaxis.set_ticks_position('none')
    ax.yaxis.set_ticks_position('none')
    
def remove_splines(ax, spl):
    for s in spl:
        ax.spines[s].set_visible(False)  

def modify_splines(ax, lwd, col):    
    for s in ['bottom', 'left','top','right']:
        ax.spines[s].set_linewidth(lwd)
        ax.spines[s].set_color(col)    
           
```

```python
fig, ax = base_figure()

remove_ticks(ax)
modify_splines(ax, lwd=0.75, col='0.8')
remove_splines(ax, ['top','right'])

ax.patch.set_facecolor('0.93')
ax.grid(True, 'major', color='0.98', linestyle='-', linewidth=1.0)
ax.set_axisbelow(True)   

fig.show()
```

### Define custom transformations

```python
def ggplot(ax):
    
    remove_ticks(ax)
    modify_splines(ax, lwd=0.75, col='0.8')
    remove_splines(ax, ['top','right'])
    
    ax.patch.set_facecolor('0.93')
    ax.grid(True, 'major', color='0.98', linestyle='-', linewidth=1.0)
    ax.set_axisbelow(True)   
```

```python
fig, ax = base_figure()
ggplot(ax)
fig.show()
```

### Legends

```python
def base_figure():
    
    fig, ax = plt.subplots(figsize=(6,5))
    for index, cyl in enumerate([4,6,8]):
        df = cars[cars['cyl'] == cyl]
        
        tmp = '{c} cyl'
        tmp = tmp.format(c=cyl)
        
        ax.scatter(df['wt'], 
                   df['mpg'], 
                   c=color[index], 
                   s=100, 
                   alpha=0.75, 
                   edgecolor='none',
                   label=tmp)  # adding a label
    
    return fig, ax

fig, ax = base_figure()

ax.legend(loc='best')

ggplot(ax)

fig.show()
```

```python
def nice_legend(ax):
    if ax.legend_ is not None:
        ax.legend_.get_frame().set_linewidth(0)
        ax.legend_.get_frame().set_alpha(0.5)
```

```python
fig, ax = base_figure()

ax.legend(loc='best', scatterpoints=1) # for a single point

ggplot(ax)
nice_legend(ax)

fig.show()        
```

### Kernel Density Estimates

```python
from sklearn.neighbors.kde import KernelDensity
```

```python
fig, ax = plt.subplots()
for i in range(3):
    
    data = 20.0*np.random.randn() + sigma*np.random.randn(100)
    x = np.linspace(data.min(), data.max(), 100)

    # For sklearn
    data = data.reshape(-1, 1)
    x = x.reshape(-1, 1)
    
    kde = KernelDensity(bandwidth=3).fit(data)  # you can adjust the 'bandwidth' parameter
    density = np.exp(kde.score_samples(x))
    ax.plot(x, density)

fig.show()
```

### `Scipy.stats`

```python
from scipy import stats

fig, ax = plt.subplots()
for i in range(3):
    
    x = 20.0*np.random.randn() + sigma*np.random.randn(300)
    xd = np.linspace(min(x)-10, max(x)+10, 100)
    density = stats.kde.gaussian_kde(x)
    ax.plot(xd, density(xd))
    
fig.show()
```

### `fill_between`

```python
import itertools
colors = itertools.cycle(mpl.rcParams['axes.color_cycle'])
```

```python
fig, ax = plt.subplots()
for i in range(4):
    
    x = 20.0*np.random.randn() + sigma*np.random.randn(300)
    xd = np.linspace(min(x)-10, max(x)+10, 100)
    density = stats.kde.gaussian_kde(x)
    
    ax.fill_between(xd, 0, density(xd), alpha=0.25, 
                    color=next(colors), 
                    linewidth=2)
    
fig.show()
```

### Combined `hist` and `kde`

```python
fig, ax = plt.subplots()
for i in range(2):
    
    x = 20.0*np.random.randn() + sigma*np.random.randn(300)
    xd = np.linspace(min(x)-10, max(x)+10, 100)
    density = stats.kde.gaussian_kde(x)
    c = next(colors)
    
    ax.hist(x, normed=1, alpha=0.25, color=c, histtype='stepfilled')
    ax.plot(xd, density(xd), alpha=0.75, color=c, linewidth=2)
    
fig.show()
```

### Error bars

```python
x = np.linspace(0, 10, 50)
xerr = np.random.normal(np.sin(x), 0.4)
y = np.sin(x)

fig, ax = plt.subplots()

ax.plot(x, y)
ax.errorbar(x, y, xerr, fmt='.k')

fig.show()
```

```python
fig, ax = plt.subplots()

ax.plot(x, y)
ad = abs(y-xerr)
ax.fill_between(x, y - ad, y + ad, color='0.5', alpha=0.2)

fig.show()
```

### Contour

Create a simple surface.

```python
x = np.linspace(-1, 1, 50)
y = np.linspace(-2, 2, 50)

X, Y = np.meshgrid(x, y)
z = X*X + Y*Y + X*Y
```

Default countour.

```python
fig, ax = plt.subplots()

ax.contour(x, y, z, 20)

fig.show()
```

```python
fig, ax = plt.subplots()

con = ax.contourf(x, y, z, 20)

fig.colorbar(con, ax=ax)
fig.show()
```

```python
fig, ax = plt.subplots()

tmp = ax.contour(x, y, z, 20, cmap=blues, alpha=0.9)
con = ax.contourf(x, y, z, 20, cmap=blues, alpha=1.)

fig.colorbar(con, ax=ax)
fig.show()
```
